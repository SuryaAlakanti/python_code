```python
import spacy
import pandas as pd

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Example sentences
data = {
    'sentence': [
        "The cat sits on the mat.",
        "I love programming in Python.",
        "He is running very fast.",
        "Hello, world! Welcome to coding."
    ]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Function to extract POS tags
def extract_pos_tags(sentence):
    doc = nlp(sentence)
    return [token.pos_ for token in doc]

# Add a column with POS tags
df['pos'] = df['sentence'].apply(extract_pos_tags)

# Count specific POS tags and add them as new columns
df['noun'] = df['pos'].apply(lambda x: x.count('NOUN'))
df['verb'] = df['pos'].apply(lambda x: x.count('VERB'))
df['adj'] = df['pos'].apply(lambda x: x.count('ADJ'))
df['punct'] = df['pos'].apply(lambda x: x.count('PUNCT'))

# Display the DataFrame
print(df)
```
